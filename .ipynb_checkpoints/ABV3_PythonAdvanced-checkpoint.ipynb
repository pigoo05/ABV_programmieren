### Aufgabe 1 – Klausurauswertung (`for`-Schleife)

Am Ende des Semesters liegen die Klausurergebnisse (in Punkten, max. 100) einer BWL-Grundlagenvorlesung vor. Die Notengrenzen sind wie folgt definiert:

| Punkte | Note |
|--------|------|
| 90 – 100 | 1,0 |
| 75 – 89 | 2,0 |
| 60 – 74 | 3,0 |
| 50 – 59 | 4,0 |
| < 50 | 5,0 (nicht bestanden) |

**a)** Iterieren Sie mit einer `for`-Schleife über die Liste `ergebnisse` und geben Sie für jede:n Studierende:n den Namen und die zugehörige Note aus.

**b)** Berechnen Sie innerhalb derselben Schleife den Durchschnitt aller Punkte und geben Sie ihn nach der Schleife aus.

**c)** Zählen Sie, wie viele Studierende die Klausur **nicht bestanden** haben (< 50 Punkte), und geben Sie die Anzahl aus.

<details>
<summary><b>Tipp zu a) – Note bestimmen</b></summary>

Verwenden Sie eine `if/elif/else`-Kette innerhalb der Schleife, um die Punktzahl in eine Note umzuwandeln.
```python
for name, punkte in ergebnisse:
    if punkte >= 90:
        note = "1,0"
    elif ...
```
</details>

<details>
<summary><b>Tipp zu b) – Durchschnitt</b></summary>

Legen Sie **vor** der Schleife eine Variable `gesamt = 0` an. Addieren Sie in jedem Schleifendurchlauf die aktuelle Punktzahl. Teilen Sie anschließend durch `len(ergebnisse)`.
</details>

<details>
<summary><b>Tipp zu c) – Zähler</b></summary>

Legen Sie analog einen Zähler `nicht_bestanden = 0` vor der Schleife an und erhöhen Sie ihn mit `+= 1`, wenn die Bedingung zutrifft.
</details>

````python
ergebnisse = [
    ("Anna Bauer",      82),
    ("Ben Schulz",      47),
    ("Clara Hoffmann",  95),
    ("Daniel Meier",    63),
    ("Elena Fischer",   55),
    ("Felix Wagner",    71),
    ("Greta Koch",      88),
    ("Hans Müller",     38),
]

# Ihre Lösung hier:
summe = 0
nicht_bestanden = 0

for name, punkte in ergebnisse:
    if 90 <= punkte <= 100:
        note = 1.0
    elif 75 <= punkte <= 89:
        note = 2.0
    elif 60 <= punkte <= 74:
        note = 3.0
    elif 50 <= punkte <= 59:
        note = 4.0
    else:
        note = 5.0
        nicht_bestanden += 1

    print(name, "hat die Note", note)
    summe += punkte

durchschnitt = summe / len(ergebnisse)

print("Durchschnitt:", durchschnitt)
print("Nicht bestanden:", nicht_bestanden)
````

### Aufgabe 4 – Break-Even-Analyse

Im Rahmen einer Unternehmensgründung soll eine Funktion zur Break-Even-Analyse implementiert werden. Der Break-Even-Punkt (BEP) gibt an, ab welcher Absatzmenge ein Unternehmen seine Gesamtkosten durch Erlöse deckt:

$$BEP = \frac{\text{Fixkosten}}{\text{Preis} - \text{variable Stückkosten}}$$

**a)** Schreiben Sie eine Funktion `break_even(fixkosten, preis, variable_kosten)`, die den BEP (gerundet auf ganze Stück) zurückgibt. Geben Sie eine Fehlermeldung aus und geben Sie `None` zurück, falls `preis <= variable_kosten` (kein positiver Deckungsbeitrag).

**b)** Rufen Sie die Funktion mit folgenden Szenarien auf und geben Sie das Ergebnis jeweils als aussagekräftigen Satz aus:
- Fixkosten: 50.000 €, Preis: 25 €, variable Kosten: 10 €
- Fixkosten: 12.000 €, Preis: 8 €, variable Kosten: 8 € *(Fehlerfall)*
- Fixkosten: 80.000 €, Preis: 120 €, variable Kosten: 45 €

<details>
<summary><b>Tipp zu a) – Formel umsetzen</b></summary>

```python
import math

def break_even(fixkosten, preis, variable_kosten):
    if preis <= variable_kosten:
        print("Fehler: Kein positiver Deckungsbeitrag.")
        return None
    bep = fixkosten / (preis - variable_kosten)
    return math.ceil(bep)  # aufrunden auf ganze Stück
```
</details>

<details>
<summary><b>Tipp zu b) – None prüfen</b></summary>

Fangen Sie den Fehlerfall ab, bevor Sie das Ergebnis ausgeben:
```python
bep = break_even(...)
if bep is not None:
    print(f"Break-Even wird bei {bep} Stück erreicht.")
```
</details>

### Aufgabe 5 – Kursverwaltung

Die Universität möchte ihr Kursangebot digital verwalten. Erstellen Sie eine Klasse `Kurs` mit folgenden Vorgaben:

**Attribute (werden bei der Erzeugung übergeben):**
- `titel` – Name der Lehrveranstaltung
- `dozent` – Name der Lehrperson
- `max_teilnehmer` – maximale Teilnehmerzahl

**Attribute (werden automatisch initialisiert):**
- `teilnehmer` – zunächst leere Liste

**Methoden:**
- `anmelden(name)` – fügt einen/eine Studierende:n zur Liste hinzu; gibt eine Fehlermeldung aus, wenn die maximale Teilnehmerzahl bereits erreicht ist
- `abmelden(name)` – entfernt eine Person aus der Liste; gibt eine Meldung aus, wenn die Person nicht angemeldet ist
- `display()` – gibt eine Übersicht des Kurses aus (Titel, Dozent:in, aktuelle Belegung)

**a)** Implementieren Sie die Klasse und führen Sie anschließend den Testcode aus.

**b)** Machen Sie alle Attribute privat (Name-Mangling mit `__`) und passen Sie die Methoden entsprechend an.

<details>
<summary><b>Tipp zu a) – Klassenstruktur</b></summary>

```python
class Kurs:
    def __init__(self, titel, dozent, max_teilnehmer):
        self.titel = titel
        self.dozent = dozent
        self.max_teilnehmer = max_teilnehmer
        self.teilnehmer = []
```
</details>

<details>
<summary><b>Tipp – anmelden() implementieren</b></summary>

```python
def anmelden(self, name):
    if len(self.teilnehmer) >= self.max_teilnehmer:
        print(f"Kurs voll – {name} konnte nicht angemeldet werden.")
    else:
        self.teilnehmer.append(name)
        print(f"{name} wurde erfolgreich angemeldet.")
```
</details>

<details>
<summary><b>Tipp zu b) – private Attribute</b></summary>

Ersetzen Sie `self.titel` durch `self.__titel` usw. Innerhalb der Klasse verwenden Sie dann ebenfalls `self.__titel`.

- **Überschriebene** `display()`-Methode – ruft das `display()` der Elternklasse auf und gibt zusätzlich die Plattform und den Link aus

**a)** Erstellen Sie die Klasse `OnlineKurs` durch Vererbung von `Kurs`.

**b)** Führen Sie den Testcode aus und überprüfen Sie die Ausgabe.

<details>
<summary><b>Tipp zu a) – Vererbungssyntax</b></summary>

```python
class OnlineKurs(Kurs):

        super().__init__(titel, dozent, max_teilnehmer)
```

`super().__init__(...)` ruft den Konstruktor der Elternklasse auf – Sie müssen die bereits definierten Attribute nicht erneut zuweisen.
</details>

<details>
<summary><b>Tipp – get_link() umsetzen</b></summary>

String-Methoden helfen beim Formatieren des Links:
```python
def get_link(self):
    titel_slug = self.titel.replace(" ", "-").lower()
    return f"https://{self.plattform.lower()}.example.com/{titel_slug}"
```
Wenn `titel` privat ist (aus Aufgabe 5b), müssen Sie ggf. eine Property oder einen Getter verwenden.
</details>

<details>
<summary><b>Tipp – display() überschreiben</b></summary>

```python
def display(self):
    super().display()                        # Ausgabe der Elternklasse
    print(f"Plattform: {self.plattform}")
    print(f"Link: {self.get_link()}")
```
</details>

### Hausaufgabe – Kassensystem mit Tageskasse

Implementieren Sie ein einfaches Kassensystem für einen kleinen Hochschulcafé-Stand.

Das System soll Bestellungen entgegennehmen, Preise berechnen und eine Tagesabrechnung ausgeben.

**a) Funktion `berechne_preis(artikel, menge)`**

Schreiben Sie eine Funktion, die den Gesamtpreis für eine Bestellung berechnet.
Verwenden Sie folgendes Preisverzeichnis:

| Artikel | Preis |
|---------|-------|
| `"Kaffee"` | 1,50 € |
| `"Tee"` | 1,20 € |
| `"Wasser"` | 0,80 € |
| `"Snack"` | 2,00 € |

Die Funktion soll `None` zurückgeben und eine Meldung ausgeben, wenn der Artikel unbekannt ist.

**b) Klasse `Tageskasse`**

Erstellen Sie eine Klasse `Tageskasse` mit folgenden Anforderungen:

**Attribute:**
- `umsatz` – Gesamtumsatz des Tages, startet bei `0.0`
- `bestellungen` – leere Liste, speichert jede Bestellung als Dictionary `{"artikel": ..., "menge": ..., "betrag": ...}`

**Methoden:**
- `bestellen(artikel, menge)` – nutzt `berechne_preis()`, fügt die Bestellung zur Liste hinzu und addiert den Betrag auf `umsatz`. Unbekannte Artikel werden nicht hinzugefügt.
- `tagesabrechnung()` – gibt eine formatierte Übersicht aller Bestellungen und den Gesamtumsatz aus.

**c) Schleife über Bestellliste**

Verarbeiten Sie die folgende Bestellliste automatisch mit einer `for`-Schleife und rufen Sie danach `tagesabrechnung()` auf:

```python
bestellungen_des_tages = [
    ("Kaffee", 3),
    ("Snack", 2),
    ("Tee", 5),
    ("Pizza", 1),   # unbekannter Artikel
    ("Wasser", 4),
    ("Kaffee", 7),
]
```

<details>
<summary><b>Tipp zu a) – Preisliste als Dictionary</b></summary>

Legen Sie das Preisverzeichnis als Dictionary an und prüfen Sie mit `if artikel not in preise`:
```python
def berechne_preis(artikel, menge):
    preise = {"Kaffee": 1.50, "Tee": 1.20, "Wasser": 0.80, "Snack": 2.00}
    if artikel not in preise:
        print(f"Unbekannter Artikel: {artikel}")
        return None
    return preise[artikel] * menge
```
</details>

<details>
<summary><b>Tipp zu b) – bestellen() implementieren</b></summary>

```python
def bestellen(self, artikel, menge):
    betrag = berechne_preis(artikel, menge)
    if betrag is not None:
        self.bestellungen.append({"artikel": artikel, "menge": menge, "betrag": betrag})
        self.umsatz += betrag
```
</details>
