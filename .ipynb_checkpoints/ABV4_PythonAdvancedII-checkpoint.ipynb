## Teil 1 - Funktionen & Indexierung

### Aufgabe 1 - Kundendaten bereinigen

Ein Shop hat rohe Kundendaten. Reparieren Sie den Code so, dass:

1. alle Namen bereinigt (`strip`, `title`) werden,
2. nur gueltige Alterswerte in den Durchschnitt eingehen,
3. keine ungueltige Indexierung oder falsche Ausnahmebehandlung mehr auftritt.

Am Ende soll der bereinigte Name und das Durchschnittsalter ausgegeben werden.

**Erwartete Ausgabe:**
```
Anna
Durchschnittsalter: 23.50
```

````python
raw_users = [
    {"name": " anna ", "age": "22"},
    {"name": "BEN", "age": "N/A"},
    {"name": "clara", "age": "25"}
]

def clean_name(user):
    return user["name"].strip().title()

def average_age(users):
    total = 0
    count = 0

    for i in range(len(users)):
        try:
            age = int(users[i]["age"])
            total += age
            count += 1
        except ValueError:
            pass

    return total / count

print(clean_name(raw_users[0]))
print(f"Durchschnittsalter: {average_age(raw_users):.2f}")
````

## Teil 2 - Schleifenlogik

### Aufgabe 2 - Lagerpruefung mit while und for

Die Funktion soll Nachbestellungen erzeugen, wenn der Bestand unter `mindestbestand` faellt.

Reparieren Sie den Code so, dass:
- die while-Schleife nicht haengen bleibt,
- die while-Schleife alle Produkte korrekt verarbeitet,
- die Ergebnisliste korrekt aufgebaut wird.

**Erwartete Ausgabe:**
```
[{'artikel': 'Kaffee', 'menge': 3}, {'artikel': 'Wasser', 'menge': 4}]
```

````python
lager = [
    {"artikel": "Kaffee", "bestand": 2},
    {"artikel": "Tee", "bestand": 9},
    {"artikel": "Wasser", "bestand": 1}
]

def nachbestellung(lagerdaten, mindestbestand=5):
    to_order = []
    i = 0

    while i < len(lagerdaten):
        eintrag = lagerdaten[i]
        if eintrag["bestand"] < mindestbestand:
            menge = mindestbestand - eintrag["bestand"]
            to_order.append({"artikel": eintrag["artikel"], "menge": menge})
        i += 1

    return to_order

print(nachbestellung(lager))
````

## Teil 3 - Klassen & Kapselung

### Aufgabe 3 - Klasse Konto debuggen

Die Klasse soll folgende Regeln korrekt umsetzen:
- Name und Startsaldo werden sauber initialisiert,
- negative Einzahlungen sind nicht erlaubt,
- private Attribute werden nicht direkt von aussen bearbeitet.

Der Testcode unten soll danach ohne Fehler laufen.

**Erwartete Ausgabe:**
```
Kontostand: 125
```

````python
class Konto:
    def __init__(self, owner, saldo=0)
        self.owner = owner
        self.__saldo = "0"

    def einzahlen(self, betrag):
        if betrag < 0:
            print("Negative Einzahlung nicht erlaubt")
        self.__saldo =+ betrag

    def stand(self):
        return self.__saldo

konto = Konto("Mara", 100)
konto.einzahlen(25)
konto.einzahlen(-200) # sollte nicht erlaubt sein
print(f"Kontostand: {konto.stand()}")
````

## Teil 4 - Vererbung

### Aufgabe 4 - OnlineKurs und PraesenzKurs

Reparieren Sie die Vererbung so, dass beide Kursarten funktionieren.

Ziel:
- Basisklasse initialisiert Titel und Teilnehmende,
- Unterklasse `OnlineKurs` nutzt `super()` korrekt,
- Methoden werden mit richtiger Signatur ueberschrieben,
- Link-Ausgabe funktioniert robust.

**Erwartete Ausgabe:**
```
Python Advanced: 2 Teilnehmende | Plattform: Zoom
https://zoom.example.com/python-advanced
```

````python
class Kurs:
    def __init__(self, titel):
        self.titel = titel
        self.teilnehmer = []

    def anmelden(self, name):
        self.teilnehmer.append(name)

    def info(self):
        return f"{self.titel}: {len(self.teilnehmer)} Teilnehmende"

class OnlineKurs(Kurs):
    def __init__(self, titel, plattform):
        self.plattform = plattform

    def get_link(self):
        return "https://" + self.plattform.lower + 
 + self.titel.replace(" ", "-").lower()

    def info():
        return super().infos() + f" | Plattform: {self.plattform}"

ok = OnlineKurs("Python Advanced", "Zoom")
ok.anmelden("Anna")
ok.anmelden("Ben")
print(ok.info())
print(ok.get_link())
````

## Teil 5 - Gemischter Debug-Fall

### Aufgabe 5 - Warenkorb auswerten

Die Auswertung soll fuer jeden Warenkorb den Gesamtwert bestimmen und am Ende den Durchschnitt aller Warenkoerbe berechnen.

Reparieren Sie den Block so, dass:
- Syntaxfehler verschwinden,
- Indexierung robust ist,
- Schleifenvariablen korrekt genutzt werden,
- keine falschen Zwischenwerte den Durchschnitt verfalschen.

**Erwartete Ausgabe:**
```
Warenkorb 1: 5.00 Euro
Warenkorb 2: 3.60 Euro
Warenkorb 3: 4.70 Euro
Durchschnitt: 4.43 Euro
```

````python
warenkoerbe = [
    [("Kaffee", 2, 1.5), ("Snack", 1, 2.0)],
    [("Tee", 3, 1.2)],
    [("Wasser", 4, 0.8), ("Kaffee", 1, 1.5)]
]

def warenkorb_wert(korb):
    total = 0
    for i in range(len(korb)):
        artikel = korb[i + 1][1]
        menge = korb[i][2]
        preis = korb[i][3]
        total = menge * preis
    return round(total,2)

gesamt = 0
for idx in range(len(warenkoerbe))
    wert = warenkorb_wert(warenkoerbe[idx])
    print(f"Warenkorb {idx + 1}: {wert} Euro")
    gesamt += wert

print(f"Durchschnitt: {gesamt / len(warenkoerbe):.2f} Euro")
````
