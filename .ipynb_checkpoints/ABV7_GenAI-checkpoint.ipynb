Schreiben Sie Kommentare, die das **Warum** erklaeren, nicht das **Was**:

```python
# Guter Kommentar: Akkumuliert Einnahmen pro Kategorie fuer spaetere Prozentberechnung

```
</details>

````python
# Dieser Code ist absichtlich ohne Kommentare. Ihre Aufgabe ist es, ihn zu lesen und zu verstehen.
# Veraendern Sie diesen Block nicht – er dient nur als Lesematerial.

from collections import defaultdict

def process(data, mode="net"):
    key = "amount_net" if mode == "net" else "amount_gross"

    buckets = defaultdict(lambda: {"in": 0.0, "out": 0.0, "n": 0})

    for row in data:
        cat = row.get("category", "?")
        val = float(row.get(key, 0) or 0)
        btype = row.get("booking_type", "")

        if btype == "revenue":
            buckets[cat]["in"] += val
        elif btype == "expense":
            buckets[cat]["out"] += val
        buckets[cat]["n"] += 1

    result = []
    for cat, vals in buckets.items():
        net = vals["in"] - vals["out"]
        result.append({
            "category": cat,
            "in": round(vals["in"], 2),
            "out": round(vals["out"], 2),
            "net": round(net, 2),
            "n": vals["n"],
            "ratio": round(vals["in"] / vals["out"], 4) if vals["out"] > 0 else None
        })

    return sorted(result, key=lambda x: -x["net"])


records = df.to_dict(orient="records")
result = process(records)
for r in result[:5]:
    print(r)
````

</details>

````python
# Aufgabe 3 - Fehlerhafter Code (NICHT VERAENDERN – nur lesen und analysieren)

def erstelle_quartalsbericht(df):
    bericht = ()                              
    df["booking_date"] = pd.to_datetime(df["booking_date"])
    df["quartal_nr"] = df["booking_date"].dt.quarter

    for q in df["quartal_nr"].unique():
        teilmenge = df[df["quartal_nr"] == q]

        einnahmen = teilmenge[teilmenge["booking_type"] == "revenue"]["amount_net"].sum()
        ausgaben  = teilmenge[teilmenge["booking_type"] == "expense"]["amount_net"].sum()

        gewinn = einnahmen + ausgaben                        
        marge  = gewinn / ausgaben * 100                     

        if gewinn < 0:                                       
            status = "Gewinn"
        else:
            status = "Verlust"

        bericht.append({                                    
            "quartal":   f"Q{q}",
            "einnahmen": round(einnahmen, 2),
            "ausgaben":  round(ausgaben, 2),
            "gewinn":    round(gewinn, 2),
            "marge_pct": round(marge, 2),
            "status":    status,
        })

    return pd.DataFrame(bericht)


bericht_df = erstelle_quartalsbericht(df.copy())
display(bericht_df)
````

<details>
<summary><b>Tipp zu 2) - Was ist ein Edge Case?</b></summary>

Ein Edge Case ist ein Grenzfall, der in normalen Tests oft nicht beruecksichtigt wird:
- Was passiert, wenn `revenue = 0`?
- Was passiert bei negativen Werten?
- Was passiert, wenn `expenses > revenue`?
- Was passiert bei sehr kleinen oder sehr grossen Zahlen?
</details>

<details>
<summary><b>Tipp zu 3) - Prompt mit Kontext</b></summary>

Geben Sie der KI sowohl die Funktion als auch Ihre bestehenden Tests:
*"Hier ist meine Python-Funktion: [Code]. Hier sind meine Tests: [Tests]. Welche Edge Cases fehlen noch?"*
</details>

<details>
<summary><b>Tipp zu 5) - Fehler in der Funktion?</b></summary>

Schauen Sie sich an, was die Funktion bei `revenue = 0` oder bei `expenses` groesser als `revenue` zurueckgibt.
Ist das Verhalten so wie erwartet? Falls nicht, was sollte die Funktion tun?
</details>

````python
def kpi_score(revenue, expenses):
    if revenue == 0:
        return None
    margin = (revenue - expenses) / revenue
    if margin > 0.3:
        tier = "A"
    elif margin > 0.1:
        tier = "B"
    else:
        tier = "C"
    return {"margin": round(margin, 4), "tier": tier}


# --- Vorhandene Tests ---

def test_normal_fall():
    result = kpi_score(10000, 6000)
    assert result is not None
    assert result["margin"] == 0.4
    assert result["tier"] == "A"
    print("test_normal_fall: OK")

def test_revenue_null():
    result = kpi_score(0, 5000)
    assert result is None
    print("test_revenue_null: OK")


test_normal_fall()
test_revenue_null()
````
