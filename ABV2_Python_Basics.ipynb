### Aufgabe 7 – Zu spät zur Vorlesung?

**a)** Generieren Sie eine zufällige Minutenzahl zwischen 0 und 200 (`randint` aus dem Modul `random`). Legen Sie anhand der folgenden Regeln mit `if / elif / else` fest, welche Meldung ausgegeben wird:

| Minuten | Ausgabe |
|---------|--------|
| 0 | Sie sind pünktlich |
| 1 – 15 | Noch im akademischen Viertel |
| 16 – 30 | Noch zur ersten halben Stunde geschafft |
| 31 – 100 | Viel, aber nicht alles verpasst |
| 101 – 129 | Fast alles verpasst |
| ≥ 130 | Alles verpasst |

**b)** Zusätzlich kann es vorkommen, dass die Lehrperson nicht erscheint. Legen Sie eine Variable an, die zufällig `True` oder `False` enthält (`choice()` aus `random`). Ergänzen Sie folgende Sonderfälle:

- Lehrperson fehlt & Sie pünktlich → *"Sie haben sich umsonst beeilt"*
- Lehrperson erscheint & Sie pünktlich → *"Sie haben es geschafft!"*
- Lehrperson fehlt & Sie zu spät → *"Sie haben nichts verpasst – aber sind umsonst gekommen"*

Geben Sie die zufällig generierten Werte zur Überprüfung mit aus.

<details>
<summary><b>Tipp 1 – Imports & Zufallswerte</b></summary>

```python
from random import randint, choice
minuten = randint(0, 200)
prof_fehlt = choice([True, False])
```
</details>

<details>
<summary><b>Tipp 2 – Struktur</b></summary>

Beginnen Sie mit `if minuten == 0`, gefolgt von mehreren `elif`-Blöcken und einem abschließenden `else`. Zeichnen Sie sich den Control-Flow auf, bevor Sie tippen.
</details>

<details>
<summary><b>Tipp 3 – Teil b) einbauen</b></summary>

Den Sonderfall "Pünktlich" können Sie mit einem **verschachtelten If** innerhalb des `if minuten == 0`-Blocks lösen:
```python

    else:
        print("Sie haben es geschafft!")
```
